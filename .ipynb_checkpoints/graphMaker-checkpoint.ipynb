### we used awk to convert the data printed into a csv file

#### the code is as follows 

```awk
BEGIN	{
    while (getline < ARGV[1]) {
        if ($1 == "Currently") {
            print "-1,\n"
        }
        # At time: 57$3 Idle/Thinking : 0$6 Requesting : 5$9 Waiting : 0$12 Eating : 0$15
        else {
            if ($1 == "Amount") {
                print ""
            }
            else {
                print $3,",",$6,",",$9,",",$12,",",$15,"\n"
            }
        }
    }
}
```

````python
! awk -f csvMaker.awk original.txt > csv1.csv
````

````python
! awk -f csvMaker.awk extraChop.txt > csv2.csv
````

### We are using MatplotLib to make our graphs

[link to the plugin and graph we want to use](https://matplotlib.org/stable/gallery/lines_bars_and_markers/barchart.html#sphx-glr-gallery-lines-bars-and-markers-barchart-py)

````python
original_rows = []

with open('csv1.csv', 'r') as csvfile:
    raw = csv.reader(csvfile, delimiter=',')
    for row in raw:
        if len(row) != 0:
            # remove the deadlocks as they can be seen during epochs where 5 are waiting
            if row[1] != '': 
                for i in range(len(row)):
                    row[i] = float(row[i])
                original_rows.append(row)
````

````python
for row in original_rows:
    print(row)
````

````python
extraChop_rows = []

with open('csv2.csv', 'r') as csvfile:
    raw = csv.reader(csvfile, delimiter=',')
    for row in raw:
        if len(row) != 0:
            # remove the deadlocks as they can be seen during epochs where 5 are waiting
            if row[1] != '': 
                for i in range(len(row)):
                    row[i] = float(row[i])
                extraChop_rows.append(row)
````

````python
for row in extraChop_rows:
    print(row)
````

````python
# Plot for the Original Algorithm

time = []
idle = []
requesting = []
waiting = []
eating = []

for row in original_rows:
    time.append(row[0])

for row in original_rows:
    idle.append(row[1])
    
for row in original_rows:
    requesting.append(row[2])
    
for row in original_rows:
    waiting.append(row[3])

for row in original_rows:
    eating.append(row[4])

fig, ax = plt.subplots()

line1, = ax.plot(time, idle, label='idle')
line2, = ax.plot(time, requesting, label='requesting')
line3, = ax.plot(time, waiting, label='waiting')
line4, = ax.plot(time, eating, label='eating')

ax.set(xlabel='Time', ylabel='Number of Philosophers')
ax.grid()

ax.legend()
plt.show()

````

````python
# Plot for the Algorithm with an extra chopstick

time = []
idle = []
requesting = []
waiting = []
eating = []

for row in extraChop_rows:
    time.append(row[0])

for row in extraChop_rows:
    idle.append(row[1])
    
for row in extraChop_rows:
    requesting.append(row[2])
    
for row in extraChop_rows:
    waiting.append(row[3])

for row in extraChop_rows:
    eating.append(row[4])

fig, ax = plt.subplots()

line1, = ax.plot(time, idle, label='idle')
line2, = ax.plot(time, requesting, label='requesting')
line3, = ax.plot(time, waiting, label='waiting')
line4, = ax.plot(time, eating, label='eating')

ax.set(xlabel='Time', ylabel='Number of Philosophers')
ax.grid()

ax.legend()
plt.show()

````
