### we used awk to convert the data printed into a csv file

#### the code is as follows 

```awk
BEGIN	{
    while (getline < ARGV[1]) {
        if ($1 == "Currently") {
            print "-1,\n"
        }
        # At time: 57$3 Idle/Thinking : 0$6 Requesting : 5$9 Waiting : 0$12 Eating : 0$15
        else {
            print $3,",",$6,",",$9,",",$12,",",$15,"\n"
        }
    }
}
```

````python
! awk -f csvMaker.awk original.txt > csv1.csv
````

````python
! awk -f csvMaker.awk extraChop.txt > csv2.csv
````
